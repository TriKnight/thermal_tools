# 2. Plot MLX90640 data form CSV file

```python
# Load your data
raw_thermal_data = pd.read_csv('./data/thermal_data0.csv')
```

```python
#show your data, the MLX90640 data has size 768 pixels from P000 to P767 
raw_thermal_data.head()
```

```python
# Drop the Time and RT value
raw_thermal= raw_thermal_data.drop(["Time", "RT"], axis=1)
raw_thermal.head()
```

```python
#Convert data to numpy
raw_thermal = raw_thermal.to_numpy()
```

```python
# Reshape data
raw_thermal = raw_thermal.reshape((-1, 24, 32))
```

```python
#Shape of the raw data
raw_thermal.shape
```

```python
#Normalize the data
v = raw_thermal
raw_data = (v - v.min()) / (v.max() - v.min())
```

```python
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(raw_data[i], cmap='jet')
plt.show()
```
